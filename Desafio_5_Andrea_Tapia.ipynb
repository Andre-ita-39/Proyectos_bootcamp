b. Generamos una tabla que contenga la unión de las 3 tablas utilizando sufijos para las columnas con nombres iguales. Para ello debemos unir primero dos tablas y después unir el resulado con la tercera tabla:

- Primero se unirá la tabla *subjects* con *incidents* formando la tabla *subjects_incidents_df*. 

- Segundo de unirá la tabla *subjects_incidents_df* con *officers*, para formar finalmente la tabla **merged_df**.

La unión de las tablas se hacen mediante la columna **case_number** que es la columna que tienen en común las 3 tablas.

En el parámetro **how** usamos el valor **outer** porque necesitamos todos los datos contenidos en las 3 tablas.

-> Hacemos el uso del prámetro **suffixes** que se usa cuando se fusionan dos DataFrames que tienen columnas con el mismo nombre. Su uso es diferenciar estas columnas en el DataFrame resultante. Este parámetro toma una tupla de dos strings que serán añadidos como sufijos a las columnas de cada DataFrame. La tupla *('_df1', '_df2')* especifica los sufijos siendo el primer string '_df1' aplicado a las columnas duplicadas de df1 (por ejemplo subjects_df), y el segundo string '_df2' aplicado a las columnas duplicadas de df2 (por ejemplo incidents_df).

```python
# Unir subjects con incidents

subjects_incidents_df = pd.merge(subjects_df, incidents_df, on='case_number', how='outer', suffixes=('_subject', '_incident'))

# Unir el resultado anterior con officers

merged_df = pd.merge(subjects_incidents_df, officers_df, on='case_number',  how='outer', suffixes=('_subject', '_officer'))

# Mostrar la tabla unida

merged_df
```

d. Generar columnas de salario mínimo y máximo estimados. Para ello:

- Usamos **apply()** para aplicar una función a cada elemento de la columna *Salary Estimate*.

- **lambda** define una función que toma un argumento **x**.

- **x.split("-")**: *split("-")* divide la cadena **x** en una lista de dos elementos, separando por el carácter "-"-". 

- **[float(i.strip()) for i in x.split("-")]**: Esta es la función que va a aplicar *lambda*, ésta recorre cada elemento **i** en la lista de *x.split("-")*, además quita espacios en blanco alrededor de **i** con **strip()**, y convierte **i** a un float.

- **[0]** selecciona el primer elemento de la lista resultante, que es el salario mínimo.

- Para **Salario Máximo Estimado** usamos el mismo código pero en vez de utilizar 0 utilizamos **[1]**, ya que selecciona el segundo elemento de la lista, que corresponde al salario máximo.



```python
# Crear la columna Salario Mínimo Estimado
trabajos_df["Salario Mínimo Estimado"] = trabajos_df["Salary Estimate"].apply(lambda x: [float(i.strip()) for i in x.split("-")][0])

#Crear la columna Salario Máximo Estimado
trabajos_df["Salario Máximo Estimado"] = trabajos_df["Salary Estimate"].apply(lambda x: [float(i.strip()) for i in x.split("-")][1])

#Mostramos el DataFrame para verificar los cambios realizados hasta ahora
trabajos_df
```
