## Ejercicio 1 - *Manipulando listas*

### Desarrollo

A partir de una lista definida se realizan las siguientes acciones:

NOTA: Se agregarán funciones *print()* antes de cada requerimiendo para que se represente de forma mas ordenada los resultados requeridos.Para crear un salto de línea usamos el caracter *\n*, que permite mostrar el texto en líneas separadas.

1. **Imprimir el quinto elemento de la lista**: Para ello usamos dentro de la función *print()* la indicación de acceder al quinto elemento posterior a una cadena (str) que indica a que pertenece el valor que se mostrará. Como el índice en Python comienza desde 0, el índice que corresponde al quinto elemento será 4. Ponemos el nombre de la variable que contiene la lista y entre corchetes [] indicamos el índice al cual pertenece el elemento al que deseamos acceder. En este caso: **lista[4]**

2. **Imprime los elementos del cuarto al sexto inclusive**: Existen varias formas de representar este requerimineto, y en base a lo aprendido en el módulo escogí dos:

- *Primera forma*: Dentro de una función *print()* escribimos una cadena que explica el el resultado que se imprimirá, para indicar el rango de elementos que se pide, ponemos el nombre de la variable y entre corchetes los índices que representan el rango donde el primer valor incluye el índice inicial y el segundo valor excluye el índice final, por lo que si queremos incluir el sexto elemento debemos poner un índice más del que le corresponde. 
En este caso: **lista[3:6]**, siendo 3 el índice del cuarto elelento y 6 el índice del séptimo elemento el cual no se inclurirá, sino que incluirá el sexto. El resultado se imprime como una lista.

- *Segunda forma*: Se crea una variable (elementos_seleccionados) donde se seleccionan los tres elementos específicos en una lista (los elementos van dentro de corchetes), para ello separamos por comas los elementos que queremos poniendo el nombre de la lista inicial y entre corchetes el número del índice que queremos seleccionar. En este caso: **elementos_seleccionados = [lista[3], lista[4], lista[5]]**. Posteriormene se crea un ciclo *for* que itererá sobre los elementos de la variable creada pudiedo así imprimir cada elemento por separado, no en forma de lista como la opción anterior.

3. **Elimina el elemento en la posición 3**: El elemento de la posición 3 posee el índice 2, para poder eliminarlo usamos la palabra reservada **del** seguido del nombre de la varible que contiene la lista y entre corchetes el índice del elemento que deseamos eliminar. En este caso: **del lista[2]**. Para representar la nueva lista con el valor eliminado se usa *print()* con una cadena que indica que hay una nueva lista.

4. **Agrega dos nuevos elementos a la lista**: Para agregar elementos al final de la lista ya creada usamos el método *.append()* con la siguiente sintaxis: nombre_lista.append(elemento_nuevo). En este caso: **lista.append("arandano")** y **lista.append("frutilla")**. Para imprimir la nueva lista usamos el mismo *print()* que en el requerimiento anterior.

5. **Ordena la lista alfabéticamente**: Para ordenar los elelentos de una lista usamos el método *.sort()* con la siguiente sintaxis: nombre_lista.sort(). En este caso escogí un ciclo *for* al igual que en el requerimiento 2 para poder representar de una forma visual más ordenada el resultado de este método. 

6. **Genera un diccionario usando lista1 como llaves y lista como valores**: Primero creamos la variable *lista1* que contiene la nueva lista para poder crear el diccionario. Después se crea la variable *diccionario* que contiene la palabra reservada **dict** para crear un diccionario y la palabra reservada **zip** para emparejar los elementos de las dos listas de forma que cada tupla que resulta tenga un elemento de cada lista en la misma posición, como queremos emparejar la **lista1** como *keys* y la **lista** como *valores* la sintaxis es la siguiente: *diccionario = dict(zip(lista1, lista))*. Finalmente usamos *print()* con una pequeña cadena para mostrar el diccionario que hemos creado.


```python
#Usaremos la siguiente lista:

lista = ["manzana", "pera", "naranja", "plátano", "kiwi", "uva", "piña",
"mango", "sandía", "melón"]

# 1) Imprimimos el 5to elemento de la lista

print("Requerimiento 1\n")

print(f"El quinto elemento de la lista es: {lista[4]}\n")


# 2) Imprimimos del cuarto al sexto elemento inclusive. Se imprime dos formas diferentes de obtener el resultao:

print("Requerimiento 2\n")

# Primera forma

print("Primera forma\n")

print(f"El cuarto, quinto y sexto elemento de la lista son:{lista[3:6]}\n")  

# Segunda forma 

print("Segunda forma\n")

elementos_seleccionados = [lista[3], lista[4], lista[5]]  

print("Los elementos 4to,5to y 6to son:")

for elemento in elementos_seleccionados:
    print("-",elemento)


# 3) Eliminamos el elemento en la posición 3

print("\nRequerimiento 3\n")

del lista[2]
print(f"La nueva lista es: {lista}\nHemos eliminado el elemento 'naranja'\n")



# 4) Agregamos dos nuevos elementos a la lista

print("Requerimiento 4\n")

lista.append("arandano")
lista.append("frutilla")

print(f"La nueva lista es: {lista}\n")


# 5) Ordenamos la lista alfabéticamente

print("Requerimiento 5\n")

lista.sort()

print("La lista ordenada alfabéticamente se ve así:")

for elemento in lista:
    print("-",elemento)
    

# 6) Genera un diccionario usando lista1 como llaves y lista como valores

print("\nRequerimiento 6\n")

lista1 = ["uno", "dos", "tres", "cuatro", "cinco", "seis", "siete", "ocho", "nueve", "diez"]

diccionario = dict(zip(lista1, lista))

print("Nuevo diccionario:\n", diccionario)
```

### Desarrollo

A continuación explicaré paso a paso el desarrollo del ejercicio en base a la guía entregada:

2. **Carga los datos del archivo csv en una variable llamada array, usando np.loadtxt(“alcohol_cerveza.csv”) y modifica los parámetros necesarios para leer bien el archivo**: Incluimos en la variable *array* la función *np.loadtxt* que lee el archivo y convierte su contenido en un array de NumPy. El parámetro *delimiter=","* indica que los valores en el archivo CSV estarán separados por comas. La modificación que se realiza en este caso es la ruta del archivo para poder hacer la correcta lectura de este.

3. **Calcula la media de los datos utilizando la función mean de NumPy y almacena el resultado en una variable llamada media**: Se crea la variable solicitada y usamos la función *np.mean*, la cual calcula la media o promedio de los valores en el array. Se usa la función *print* para mostrar el resultado y se reducen los decimales a 4 dígitos.

4. **Calcula la desviación estándar de los datos utilizando la función std de NumPy y almacena el resultado en una variable llamada desviacion**: Se crea la variable *desviacion* donde almacenamos la función *np.std* que calcula la desviación estándar de los valores en el array. La desviación estándar mide la dispersión de los valores respecto a la media. Usamos la función *print()* para imprimir el resultado.

5. **Calcula el límite inferior de control (LIC) utilizando la fórmula: media - (3 * desviacion) y almacena el resultado en una variable llamada lic**: Se crea la variable solicitada y se calcula el Límite Inferior de Control (LIC) restando tres veces la desviación estándar de la media. Este valor se utiliza en el control estadístico de procesos para identificar valores que están significativamente por debajo del promedio esperado. Se imprime el resultado.

6. **Calcula el límite superior de control (LSC) utilizando la fórmula: media + (3 * desviacion) y almacena el resultado en una variable llamada lsc**: Se crea la variable *lsc* y se calcula el Límite Superior de Control (LSC) sumando tres veces la desviación estándar a la media. Este valor se utiliza para identificar valores que están significativamente por encima del promedio esperado. Se imprime el resultado con la función *print()*.

7. **Filtra el array utilizando np.logical_or(condicion1, condicion2) para obtener los valores fuera de los límites de control**: Se crea la variable *fuera_de_control* que es el array que contendrá solo los valores que están fuera de los límites, para ello usamos *np.logical_or* para crear una máscara booleana que identifica los valores en el array que están fuera de los límites de control usando dos condiciones:
- Condición 1:por debajo del LIC
- Condición 2:por encima del LSC


Para el punto 7 se imprimen dos valores:
* Primero se imprimen los valores que estan fuera de los límites como un array, no como una lista de Python.
* Segundo se imprime la cantida de de valores que cumplen esta característica usando la función *len()* sobre la variable *fuera_de_control*, la cual contará la longitud del array y nos entregará la cantidad de valores.

NOTA: Se usa diferentes formas de print solamente para demostrar las diferentes formas de imprimir los datos. En todos los resultados se reducen los decimales a 4 dígitos solamente para hacer más legible el resultado.

```python
# 1.- Importamos libreria NumPy

import numpy as np

# 2.- Cargamos los datos del archivo csv

array = np.loadtxt("/Users/andrea/Desktop/DATA SCIENCE/Modulo 2-Python/python  III/alcohol_cerveza.csv", delimiter=",")  

# 3.- Calculamos la media

media = np.mean(array)
print(f"El valor de la media es: {media:.4f}")

# 4.- Calculamos la desviación estándar

desviacion = np.std(array)
print("La desviación estándar es:", round(desviacion,4))

# 5.- Calculamos LIC

lic = media - (3 * desviacion)
print(f"El Límite Inferior de Control (LIC) es: {lic:.4f}")

# 6.- Calculamos LSC

lsc = media + (3 * desviacion)
print(f"El Límite Superior de Control (LSC) es: {lsc:.4f}")

# 7.-  Filtro para obtener los valores fuera de los límites de control

fuera_de_control = array[np.logical_or(array < lic, array > lsc)]

print("Valores fuera de los límites de control son:", fuera_de_control)
print("Cantidad de valores fuera de los límites de control ->", len(fuera_de_control))
```
